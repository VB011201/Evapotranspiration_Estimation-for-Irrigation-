```python
mask=Table3_HR3['YEAR']>=1993.0 
Table3_HR3=Table3_HR3[mask]
mask2=Table3_HR3['YEAR']<=2010
Table3_HR3=Table3_HR3[mask2]
mask=Table3_HR12['YEAR']>=1993.0 
Table3_HR12=Table3_HR12[mask]
mask2=Table3_HR12['YEAR']<=2010
Table3_HR12=Table3_HR12[mask2]
mask=Radiations['YEAR']>=1993.0 
Radiations=Radiations[mask]
mask2=Radiations['YEAR']<=2010
Radiations=Radiations[mask2]
mask=Table2['YEAR']>=1993.0 
Table2=Table2[mask]
mask2=Table2['YEAR']<=2010
Table2=Table2[mask2]
```

Combinig HR_3 and HR_12 tables to one and taking the relative humidity as mean of two times

```python
Table3_HR3=Table3_HR3.drop(['HR','AW'],axis=1)
Table3_HR12=Table3_HR12.drop(['HR','AW'],axis=1)
humid=pd.DataFrame()
humid=Table3_HR3.merge(Table3_HR12,on=['YEAR','MN','DT'],how='right')
for i in range(len(humid)):
  if(humid.iloc[i][3]>humid.iloc[i][4]):
    humid.iloc[i][3],humid.iloc[i][4]=humid.iloc[i][4],humid.iloc[i][3]
humid['avg_RH']=humid[['.RH_x','.RH_y']].mean(axis=1) #.RH_x min humidity
```

Preparing final data and filling missing values using MICE

```python
Final_data=Table2.merge(humid,on=['YEAR','MN','DT'],how='left').merge(Radiations,on=['YEAR','MN','DT'],how='left')
Final_data=Final_data.drop(['RA','DZ','SN','SL','HA','TH','DS','FG','GA','W1','T','avg_RH'],axis=1)
Final=MICE.fit_transform(Final_data)
Final_data=pd.DataFrame(Final,columns=['Year','MN','DT','T_max','T_min','U_z','Rainfall','RH_min','RH_max','R_n','n_sun'])
Final_data=Final_data.drop(['Rainfall'],axis=1)
Final_data.isnull().sum()
```

Combining Year MN and DT columns and converting to date time format in order to feed data into the ET0 calculation model

```python
Final_data['Year']=Final_data['Year'].astype(int)
Final_data['MN']=Final_data['MN'].astype(int)
Final_data['DT']=Final_data['DT'].astype(int)
Final_data['Year']=Final_data['Year'].astype(str)
Final_data['MN']=Final_data['MN'].astype(str)
Final_data['DT']=Final_data['DT'].astype(str)
col=['-'.join(i) for i in zip((Final_data['Year']),(Final_data['MN']),(Final_data['DT']))]
print(col)
Final_data.insert(0,'date',col)
Final_data=Final_data.drop(['Year','MN','DT'],axis=1)
Final_data['date']=pd.to_datetime(Final_data['date']).dt.strftime('%Y-%m-%d')
Final_data.isnull().sum()
Final_data.to_csv('/content/drive/MyDrive/ProfProj/Final_Data.csv')
```
